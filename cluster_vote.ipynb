```python
import numpy as np


def PointsInCircum(r,n=100):
    return [(math.cos(2*pi/n*x)*r,math.sin(2*pi/n*x)*r) for x in xrange(0,n+1)]
```

```python
points = []
radius = [0,1,2,3]
number = [1,8,16,8]
for i in range(4):
    points = points + PointsInCircum(radius[i],number[i])

import copy
left = np.array(points)
left[:,0] = left[:,0] + 4
right = np.array(points)
right[:,0] = right[:,0]-4
out1 = np.row_stack((left,right))
candidates = np.array([(2,0),(-2,0),(0,0),(1,0),(-1,0),(3,0),(-3,0)])
if 1:
    out2 = copy.copy(out1)
    out1[:,1] = out1[:,1] - 4
    out2[:,1] = out2[:,1] + 4
    out3 = np.row_stack((out1,out2))
    out = out3
    can2 = copy.copy(candidates)
    can2[:,1] = can2[:,1] + 4
    can1 = copy.copy(candidates)
    can1[:,1] = can1[:,1] - 4
    can3 = np.row_stack((can1,can2))
    can4 = copy.copy(can3)
    can4[:,1] = can3[:,0]
    can4[:,0] = can3[:,1]
    candidates = np.row_stack((can3,can4))
else:
    out = out1

numpy.savetxt("circle2.csv", out, delimiter=",",fmt="%1.4f")
numpy.savetxt("circle2-can.csv", candidates, delimiter=",",fmt="%1.0f")
```
