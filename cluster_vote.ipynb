```python
# now let's try to pick the best candidates out of a subset

def get_score(rate,sc):
    # find the score, given ballots and a selection of candidates
    # rate and sc are inputs (sc is random)
    # score is output
    A = np.array(rate)  # ballots
    (m,n) = np.shape(A)
    seats = len(sc)
    ps = m/seats  # this penalty is an interesting thing to scale, to see what changes.
    totals = np.sum(A,0) # sum columns
    pc = ps / totals # elementwise division
    Awin = A[:,sc]  #only winners  # we could alternately use a matrix multiplication to a boolean array
    pcwin = pc[sc] 
    pr = np.dot(Awin,pcwin) # penalty for row
    pAwin = np.dot(np.reshape(pr,(m,1)), np.ones((1,seats))) # replicate column over all rows using outer product
    if 1:
        cut0 = lambda x: .5 * (x+abs(x)) # don't allow negative weights of voters. Just take them out of consideration
        # negative is not okay because it doesn't collapse the differential rating of voters
        # but maybe it is okay to keep the differential rating... not sure
        # oh, the penalty is proportional to their support of the candidate
        # so more penalty could reverse the preferences of the voter
    else:
        cut0 = lambda x: x  # not okay
        # reverses the preferences of the voter when the voter has won too many candidates.
    nA = cut0(Awin-pAwin)  
    score = sum(sum(nA))
    return score

import itertools
#real_candidates= [11,12,13,14,15,16,17]
real_candidates = range(n)
co = itertools.combinations(real_candidates, 3)# combinations of candidates
sca = []
cand2 = []
for sc in co:
    score = get_score(rate,np.array(sc))
    sca = sca + [score]
    cand2 = cand2 + [sc]
max_value = max(sca)
max_index = sca.index(max_value)
best_candidates = cand2[max_index]


fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
x= candidates[real_candidates,0]
y = candidates[real_candidates,1]
ax.scatter(x, y, color = rgb)

x= candidates[best_candidates,0]
y = candidates[best_candidates,1]
ax.scatter(x, y, color = rgb,s=100)
plt.show()

```

So basically, how this all works is the voters receive a penalty for winning a candidate.  The sum of the penalties over all voters is the same for each candidate.  (but technically, once the penalty removes a voter from consideration, because that voter has won enough candidates, the penalty for additional wins doesn't matter [although I could change this so that his preferences are reversed]).  We are balancing voter wins with candidate wins, kind of.
